```python
#homework-3_Day_4

def prnums():
    nums=list(range(100))
    for i in nums:
       
        counter=1
        if i%counter==0:
            print(i,"is a prime number.")
            counter+=1
            i+=1
        else:
            print (i,"is not a prime number")
```

```python
print(prnums())
```
